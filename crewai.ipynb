````python
import getpass
import os
from langchain_groq import ChatGroq

os.environ["GROQ_API_KEY"] = getpass.getpass()

# Initialize the model
model = ChatGroq(model="llama-3.3-70b-versatile")
````

````python
from crewai import Agent
# from crewai_tools import SerperDevTool
# Create an agent with all available parameters
writter = Agent(
    role="Senior Writter",
    goal="Analyze this summary: {project_summary}  and write a Professional document to show impress investors",
    backstory="With over 10 years of experience, "
              "you excel at writting detailed reports.",
    llm="groq/llama-3.3-70b-versatile",
    function_calling_llm=None,  # Optional: Separate LLM for tool calling
    memory=True,  # Default: True
    verbose=False,  # Default: False
    allow_delegation=False,  # Default: False
    max_iter=20,  # Default: 20 iterations
    max_rpm=None,  # Optional: Rate limit for API calls
    max_execution_time=None,  # Optional: Maximum execution time in seconds
    max_retry_limit=2,  # Default: 2 retries on error
    allow_code_execution=False,  # Default: False
    code_execution_mode="safe",  # Default: "safe" (options: "safe", "unsafe")
    respect_context_window=True,  # Default: True
    use_system_prompt=True,  # Default: True
    # tools=[SerperDevTool()],  # Optional: List of tools
    # knowledge_sources=None,  # Optional: List of knowledge sources
    # embedder_config=None,  # Optional: Custom embedder configuration
    # system_template=None,  # Optional: Custom system prompt template
    # prompt_template=None,  # Optional: Custom prompt template
    # response_template=None,  # Optional: Custom response template
    # step_callback=None,  # Optional: Callback function for monitoring
)
````

````python
Edittor = Agent(
    role="Senior Edittor",
    goal="Analyze the document written by the writter and edit it to make it appealing for investors to invest in that project",
    backstory="With over 10 years of experience, "
              "you excel at editting and proof reading reports.",
    llm="groq/llama-3.3-70b-versatile",
    function_calling_llm=None,  # Optional: Separate LLM for tool calling
    memory=True,  # Default: True
    verbose=False,  # Default: False
    allow_delegation=False,  # Default: False
    max_iter=20,  # Default: 20 iterations
    max_rpm=None,  # Optional: Rate limit for API calls
    max_execution_time=None,  # Optional: Maximum execution time in seconds
    max_retry_limit=2,  # Default: 2 retries on error
    allow_code_execution=False,  # Default: False
    code_execution_mode="safe",  # Default: "safe" (options: "safe", "unsafe")
    respect_context_window=True,  # Default: True
    use_system_prompt=True,  # Default: True
    # tools=[SerperDevTool()],  # Optional: List of tools
    # knowledge_sources=None,  # Optional: List of knowledge sources
    # embedder_config=None,  # Optional: Custom embedder configuration
    # system_template=None,  # Optional: Custom system prompt template
    # prompt_template=None,  # Optional: Custom prompt template
    # response_template=None,  # Optional: Custom response template
    # step_callback=None,  # Optional: Callback function for monitoring
)
````

````python
from crewai import Task

writter_task = Task(
    description="""
        Read the summary of different features of our project and write a detailed report
        that will impress the investors. Highlight the key benefits, unique selling points,
        and the potential market impact of the project. Make sure the report is compelling
        and well-structured to capture the investors' attention.
    """,
    expected_output="""
        A comprehensive and well-organized report that outlines the project's features,
        emphasizes its value propositions, and demonstrates how it stands out in the market.
        The report should also include an introduction, main body with details of the features,
        and a conclusion with a call-to-action for the investors.
    """,
    agent=writter
)


proof_read_task = Task(
    description="""
        Review the provided context and expand each topic into a full section for a report.
        Ensure the report is thorough, detailed, and contains all relevant information.
        Edit for clarity, coherence, and professionalism while maintaining the intent of the original content.
    """,
    expected_output="""
        A fully developed report with main topics expanded into detailed sections,
        each containing comprehensive information. The report should be well-structured,
        polished, and formatted as markdown without '```'.
    """,
    agent=Edittor,
    output_file="report.md"
)

````

````python
from crewai import Crew, Process
crew = Crew(
    agents=[writter, Edittor],
    tasks=[writter_task, proof_read_task],
    process=Process.sequential,
    verbose=True
)

crew_output = crew.kickoff(inputs={"project_summary": """Imagine you're hungry and craving a specific dish but can't remember where you last had it. You search through your apps, but nothing helps. Enter FoodFinder, your ultimate culinary assistant.FoodFinder remembers the places where you've eaten your favorite dishes and helps you locate them whenever you need. Lost track of that amazing pizza joint you visited last week? Simply ask, "Hey FoodFinder, where was that pizza place?" and you'll instantly get an answer, along with the address, directions, and even the dish you loved most.

FoodFinder is voice-activated for seamless interaction. Just say, "Hey FoodFinder, where did I have that amazing sushi?" and it will respond instantly. Forget about scrolling through old photos or restaurant apps. This app leverages natural language processing, so you don't need to follow any rigid commands — you just talk naturally!

But there's more! You can add security with a personalized activation phrase. Set a custom keyword like "Hey Chef," and next time you ask, "Where's my favorite burger?" it will only respond if you use the secret phrase: "Hey Chef, where's my favorite burger?" Your data stays private, and the app helps you with just the right touch of protection. """})


````
